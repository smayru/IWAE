```python
import numpy as np
import csv
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
N=60000  #training data size
```

```python
name1="./training_error_sampling_number_1.csv"
name5="./training_error_sampling_number_5.csv"
name10="./training_error_sampling_number_10.csv"
```

```python
data1=np.loadtxt(name1,delimiter=',')[:,1]/N
data5=np.loadtxt(name5,delimiter=',')[:,1]/N
data10=np.loadtxt(name10,delimiter=',')[:,1]/N
```

```python
plt.plot(data1,label='k=1')
plt.plot(data5,label='k=5')
plt.plot(data10,label='k=10')
plt.xlabel("Epoch")
plt.ylabel("ELBO")
plt.legend()
plt.show()
```
